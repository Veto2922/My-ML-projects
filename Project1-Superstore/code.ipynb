Postal Code => Postal Code of every Customer.

Category => Category of the product ordered.

```python
new_df = df.drop(['Order ID', 'Order Date', 'Ship Date', 'Customer ID',
                 'Postal Code', 'Product ID', 'Customer Name'], axis=1)
```

### 2)Encoding


```python
Encoded_df = pd.get_dummies(new_df)
```

### 3)split X and y

---


### 4)PCA


```python
from sklearn.decomposition import PCA
pca = PCA(n_components=50)
decom_X = pca.fit_transform(Encoded_df)
```

```python
pca.explained_variance_ratio_
```

### 4)Trian test split

---


### 5) Scaling


we will use robust scaler besouce the data containing outliers


```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
scaled_X = scaler.fit_transform(decom_X)
```

# Modeling (customer segmentation)


```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
```

```python
classes = model.fit_predict(scaled_X)
```

```python
len(classes)
```

```python
df['clusters'] = classes
```

```python
df
```

```python
sns.scatterplot(data=df, x='Sales', y='Profit', hue='clusters', palette='Set1')
```
